\subsection{Contraction}
Muscle contraction is based on the \textbf{sliding-filament theory}

\subsubsection{Contraction cycle}
$Ca^{2+}$ binds to troponin, tropomyosin moves and reveal myosin-binding sites

\begin{enumerate}
    \item Rigor is a transient state: Myosin touches actin 
    \item ATP binding dissociates myosin from actin 
    \item Myosin ATPase hydrolizes ATP to ADP, myosin reattaches to actin, ADP+$P_i$ remains bound to myosin
    \item The myosin head attaches to the actin filament
    \item ATP hydrolysis move the myosin head and the actin filament
    \item ADP is removed and myosin is attached to actin
\end{enumerate}

\subsubsection{Excitation-contraction coupling}
\textbf{Transverse tubules, T-tubules, T-tubuli, T-putket}

\begin{enumerate}
    \item Action potential moves in a motor neuron, triggers exocytosis of acetylcholine
    \item Ligand-gated channels bind ACh, open. $Na^+$ influx triggers an action potential
    \item Action potential moves along T tubules, activated voltage-sensitive DHPR (dihydropyridine receptors)
    \item DHPR activates RyR (ryanodine receptors), $Ca^{2+}$ diffueses from SR to the cytosol
    \item Contraction cycle
\end{enumerate}

\subsection{Whole skeletal muscle}
\textbf{Concentric contraction}: shortening
\newline
\noindent
\textbf{Eccentric contraction}: lengthening
\newline
\noindent
\textbf{Isometric contraction}: elastic elements stretched, the muscle stays the same length
\newline
\noindent
\textbf{Isotonic contraction}: the muscle shortens or lengthens
\newline
\noindent
\textbf{Twitch, Lihasnykäys, Twitch}: mechanical response to a single action potential
\newline
\noindent
The velocity of shortening decreases as the load increases
\newline
\noindent
\textbf{Tetanus, tetaaninen lihassupistus, tetanisk muskelkontraktion}: maximal summation of twitches, fused contraction. A sustained high calcium concentration in the cytoplasm permits summation and tetanus
\newline
\noindent
The amount of tension developed by a muscle depends on the length of the muscle at the time it is stimulated
\newline
\noindent
The amount of work of a muscle can do depends on its volume

\subsection{Muscle energetics}

\subsubsection{Sources of energy}
\begin{enumerate}
    \item Phosphagen creatine phosphate: very fast rate of ATP synthesis, small yield, fast acceleration of production
    \item Anaerobic glycolysis: high rate, moderae yield, fast acceleration
    \item Aerobic catabolism: moderate rate, high yield, slow acceleration
\end{enumerate}

\subsubsection{Types of mammalian skeletal twitch muscle fibers}
\begin{enumerate}
    \item \textbf{Slow oxidative, hitaat aerobiset lihassyyt, långsamma aeroba muskelfibrer}: slow ATPase activity, slow peak reaching, long duration of twitches, many mitochondria, high myoglobin content, red colour, Posture
    \item \textbf{Fast oxidative glycolytic, välimuotoiset lihassyyt}: fast ATPase activity, intermediate peak reaching, short duration of twitches, many mitochondria, high myoglobin, red colour, standing, walking
    \item \textbf{Fast glycolytic, Nopeat anerobiset lihassyyt}: fast ATPase activity, fast peak reaching, 
\end{enumerate}
