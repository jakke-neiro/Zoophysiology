\subsection{Homeostasis}
Homeostasis or homoeostasis (homeo- + -stasis) is the property of a system in which variables are regulated so that internal conditions remain stable and relatively constant.
Homeostasia: järjestelmän taipumus pitää yllä dynaamista tasapainoa ja häiriön sattuessa palautua tähän tasapainoon omien säätelyjärjestelmien avulla. Fysiologinen homeostasia on eliön sisäinen vakaa tila, jota lukuisat biokemialliset, hormonaaliset ja hermostolliset säätelyjärjestelmät ylläpitävät. Se on palautuva fenotyyppinen reaktiojärjestelmä, joka sisäisten säätelymekanismien avulla puskuroi eliön vaihtelevia ulkoisia olosuhteita vastaan. 
Homeostas: Homeostas, av grek. homoios (liknande) och stasis (stopp, stillastående), är ett stabilt och konstant tillstånd gentemot omgivningen i ett biologiskt system. De tre viktigaste homeostatiska mekanismerna är termoreglering, osmoreglering och reglering av blodsockernivåerna.
